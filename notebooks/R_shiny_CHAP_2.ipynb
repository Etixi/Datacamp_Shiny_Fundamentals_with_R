###**Réactivité 101**

####**1. Réactivité 101**


+ L’un des concepts les plus importants de Shiny est le concept de **réactivité**.

####**2. Principes de base de la réactivité**

+ **La réactivité, ou programmation réactive, est un élément clé de ce qui rend Shiny si puissant**. 

+ La réactivité est ce qui permet aux sorties de *réagir* automatiquement aux changements d’entrées. 

+ La réactivité dans Shiny peut sembler intimidante, mais à un niveau de base, cela signifie que lorsque la valeur d’une variable $x$ change, tout ce qui repose sur $x$ est réévalué. 

+ C’est très différent de ce à quoi vous êtes habitué dans R. 

+ Considérez ce code: $x$ est égal à 5. $y$ est égal à $x+ 1$. 

+ Alors $X$ obtient 10. Normalement, vous diriez que $y$ a la valeur de 6. 

+ Mais si $x$ et $y$ sont tous deux des variables réactives, alors $y$ serait en fait 11 parce que $y$ serait mis à jour chaque fois que $x$ change. 

+ C’est un concept très puissant qui est très utile pour la réactivité et l’interactivité de Shiny, mais il peut sembler déroutant au début.

####**3. Variables réactives**

+ Seules les variables réactives se comportent de cette façon, et dans Shiny, toutes les entrées sont automatiquement réactives.

####**Instructions**

+ On vous fournit une application Shiny contenant deux entrées numériques, num1 et num2, et une sortie texte. Votre tâche est la suivante :

````r
ui <- fluidPage(
  numericInput("num1", "Number 1", 5),
  numericInput("num2", "Number 2", 10),
  textOutput("result")
)

server <- function(input, output) {
  # Calculate the sum of the inputs
  my_sum <- reactive({
    input$num1 + input$num2
  })

  # Calculate the average of the inputs
  my_average <- reactive({
    my_sum() / 2
  })
  
  output$result <- renderText({
    paste(
      # Print the calculated sum
      "The sum is", my_sum(),
      # Print the calculated average
      "and the average is", my_average()
    )
  })
}

shinyApp(ui, server)
````

+ L'utilisateur peut faire glisser la barre le long du curseur pour choisir un nombre. 

+ Comme les entrées numériques, les curseurs ont également un minimum et un maximum, mais pour les curseurs, ces arguments sont obligatoires. 

+ Une autre grande différence entre les curseurs et les entrées numériques est que les curseurs peuvent également être utilisés pour sélectionner deux nombres à la fois. 

+ Pour ce faire, l'argument valeur doit être un vecteur de deux nombres plutôt qu'un seul nombre. 

+ Le curseur aura alors une limite inférieure et une limite supérieure, et la valeur de retour de cette entrée sera un vecteur de deux nombres.

````r
# Load the plotly package
library("plotly")

ui <- fluidPage(
  sidebarLayout(
    sidebarPanel(
      textInput("title", "Title", "GDP vs life exp"),
      numericInput("size", "Point size", 1, 1),
      checkboxInput("fit", "Add line of best fit", FALSE),
      colourInput("color", "Point color", value = "blue"),
      selectInput("continents", "Continents",
                  choices = levels(gapminder$continent),
                  multiple = TRUE,
                  selected = "Europe"),
      sliderInput("years", "Years",
                  min(gapminder$year), max(gapminder$year),
                  value = c(1977, 2002))
    ),
    mainPanel(
      # Replace the `plotOutput()` with the plotly version
      plotlyOutput("plot")
    )
  )
)

# Define the server logic
server <- function(input, output) {
  # Replace the `renderPlot()` with the plotly version
  output$plot <- renderPlot({
    # Convert the existing ggplot2 to a plotly plot
    renderPlotly({
      data <- subset(gapminder,
                     continent %in% input$continents &
                       year >= input$years[1] & year <= input$years[2])
      
      p <- ggplotly(data, aes(gdpPercap, lifeExp)) +
        geom_point(size = input$size, col = input$color) +
        scale_x_log10() +
        ggtitle(input$title)
      
      if (input$fit) {
        p <- p + geom_smooth(method = "lm")
      }
      p
    })
  })
}

shinyApp(ui = ui, server = server)
````

###**Variables réactives**

####**1. Variables réactives**
+ Vous l'avez peut-être remarqué dans les derniers exercices,

####**2. Duplication du code**
+ le code pour filtrer les données, qui fait 11 lignes, a été dupliqué trois fois. 

+ Il a été utilisé initialement dans **la fonction rendertable pour les afficher dans un tableau**, puis il a été copié dans la fonction **renderplot**, et enfin il a également été dupliqué dans le code du gestionnaire de téléchargement.

####**3. Les variables réactives réduisent la duplication du code**
+ Un problème potentiel lié à la duplication de morceaux de code est que toutes les copies du code doivent être constamment maintenues. 

+ Par exemple, si de nouvelles entrées sont ajoutées à l'application et que le code de filtrage devient plus long, vous devez penser à mettre à jour le code aux trois endroits. 

+ De même, si des bogues sont découverts et que vous devez mettre à jour le code, il faudra le faire à tous les endroits. 

+ Avec le temps, il est facile de laisser accidentellement les multiples morceaux de code devenir différents en ne les mettant pas tous à jour, ce qui peut introduire des erreurs dans votre application. 

+ Dans Shiny, lorsque le code dupliqué contient des valeurs réactives, la création d'une variable réactive est une bonne solution.

####**4. Variables réactives**
+ Pour illustrer comment les variables réactives peuvent être utilisées pour résoudre la duplication du code, supposons que le code de filtrage des données dans la fonction renderTable est dupliqué à plusieurs endroits. 

+ Pour le transformer en variable réactive, nous créons d'abord une variable, dans ce cas nous l'appellerons my-underscore-data, et nous lui assignons la fonction `reactive()`. 

+ Nous copions ensuite le morceau de code dupliqué dans la fonction réactive. 

+ Maintenant, my-underscore-data est une variable réactive qui peut être utilisée dans la fonction renderTable ou n'importe où ailleurs où le code a été dupliqué. 

+ Rappelez-vous simplement que lorsque vous accédez à une variable réactive, vous devez ajouter des parenthèses après le nom de la variable.

####**5. Mise en cache des variables réactives**
+ En plus de contribuer à réduire la duplication du code, l'utilisation des variables réactives présente quelques autres avantages importants. 

+ Les variables réactives font quelque chose appelé **mise en cache** - elles mettent leur valeur en cache. 

+ Cela signifie que les variables réactives stockent leur valeur après leur première exécution, et qu'elles n'exécutent pas leur code à nouveau, sauf si l'une de leurs dépendances a été modifiée. 

+ Cela peut être particulièrement utile lorsqu'une variable réactive prend beaucoup de temps à calculer.

####**6. Mise en cache des variables réactives**
+ Prenons l'exemple suivant : supposons qu'une fonction d'ajustement de modèle s'exécute en cinq secondes et qu'elle utilise, ou dépende, de l'une des valeurs d'entrée de l'application. 

+ Dans ce code serveur, les fonctions de rendu du tableau et du graphique utilisent le résultat de l'ajustement. 

+ Comme la fonction fit est appelée deux fois, l'exécution de ce code prend 10 secondes. 

+ Supposons maintenant qu'une variable réactive `x` soit créée pour contenir le code d'ajustement du modèle dupliqué. 

+ Maintenant, les deux fonctions de rendu peuvent utiliser cette variable réactive. 

+ Même si x est appelé deux fois, il ne s'exécute que la première fois, car la deuxième fois qu'il est appelé, il se souvient de sa valeur et ne relance pas l'ajustement du modèle. 

+ Cela signifie que ce code s'exécute en 5 secondes, ce qui représente un gain de vitesse significatif.

####**7. Les variables réactives sont paresseuses**
+ Une autre propriété utile des variables réactives est qu'elles sont **paresseuses**. 

+ Cela signifie qu'une variable réactive n'est exécutée que lorsque sa valeur est nécessaire. 

+ Pour voir comment cela peut être utile, considérez ce code. 

+ Encore une fois, nous avons une variable réactive $x$ qui prend beaucoup de temps à calculer et qui dépend d'une entrée. 

+ Cette variable réactive est utilisée dans un gestionnaire de téléchargement, ce qui signifie que chaque fois que l'utilisateur demande à télécharger le fichier, x sera appelé. 

+ L'avantage de la paresse est que grâce à elle, $x$ n'est pas recalculée à chaque fois que input-dollar-sign-num est mise à jour - ce serait très lent et une perte de temps. 

+ $x$ n'est recalculée que lorsque l'entrée change *et* sa valeur est nécessaire, c'est-à-dire seulement après que l'utilisateur ait cliqué sur le bouton de téléchargement.

###**Ajout de sources de mots**

####**1. Ajout de sources de mots**
+ Jusqu'à présent, l'application dispose d'entrées pour les paramètres facultatifs du nuage de mots, mais il n'y a aucun moyen de modifier la source des données. 

+ Nous allons découvrir deux nouveaux types d'entrées qui peuvent être utilisées pour modifier les données du nuage de mots.

####**2. Entrées de type Textarea**
+ Puisque l'argument de données de create_wordcloud est du texte, il pourrait être logique d'utiliser une entrée de texte pour capturer le texte de l'utilisateur et l'utiliser dans le nuage de mots. 

+ Mais il y aurait un problème avec cela. 

+ Les entrées de texte ne peuvent fournir qu'une ligne, ou une rangée, il est donc très peu pratique d'y entrer beaucoup de texte. 

+ L'entrée textAreaInput fournit une zone de texte plus grande, appelée textarea, à cette fin. 

+ Elle est très similaire à textinput, mais elle est idéale pour les textes longs car elle comporte plusieurs lignes et contient une barre de défilement verticale. 

+ Les textareas peuvent même avoir un nombre ajustable de rangées en spécifiant l'argument `rows`.

####**3. Entrées de fichiers - Interface utilisateur**

+ Une autre façon de fournir vos propres mots peut être de télécharger un fichier texte au lieu de taper du texte dans une zone de texte. 

+ Le chargement de fichiers dans une application Shiny se fait à l'aide de la fonction fileinput. 

+ Les entrées de fichiers ont les deux paramètres communs à toutes les autres entrées, inputId et label, ainsi que d'autres paramètres uniques que nous n'utiliserons pas. 

+ L'ajout d'une entrée fichier à l'interface utilisateur d'une application Shiny entraîne l'affichage de cette entrée.

####**4. Entrées de fichiers - Interface utilisateur**

+ Lorsque l'utilisateur clique sur le bouton Parcourir, un sélecteur de fichier apparaît et permet à l'utilisateur de sélectionner un fichier sur son ordinateur.

+ Il est possible d'autoriser l'utilisateur à sélectionner plusieurs fichiers au lieu d'un seul, ou de limiter l'utilisateur à un certain type de fichier. 

+ Vous pouvez consulter la documentation sur l'entrée des fichiers pour voir quels autres paramètres peuvent être utilisés pour personnaliser la sélection des fichiers, mais nous n'utiliserons pas ces options.

####**5. Entrées de fichiers - serveur**
+ Après que l'utilisateur ait sélectionné un fichier, celui-ci est téléchargé sur l'ordinateur qui exécute l'application Shiny, et il devient disponible pour vous dans le serveur. 

+ Mais comment accéder exactement à ce fichier ? 

+ Avec d'autres types d'entrées, chaque fois que l'utilisateur modifie l'entrée, la nouvelle valeur est disponible dans le serveur via **input$inputId**. 

+ Vous pourriez vous attendre à ce que **input$inputId** vous donne accès au fichier qui a été téléchargé, mais ce n'est pas ainsi que fonctionnent les entrées de fichiers.


####**6. Entrées de fichiers - serveur**
+ En accédant à la valeur d'entrée, vous obtiendrez un cadre de données contenant quelques informations sur le ou les fichiers sélectionnés par l'utilisateur. 

+ Chaque ligne du cadre de données correspond à un fichier. 

+ Dans ce cours, nous n'utiliserons que des entrées de fichiers qui acceptent des fichiers uniques, il n'y aura donc qu'une seule ligne. 

+ Chaque fichier comporte 4 informations : **nom, taille, type et chemin de données**. 

+ Par exemple, s'il s'agit du cadre de données renvoyé par une entrée de fichier, cela signifie que l'utilisateur a sélectionné un fichier en texte brut appelé monfichier-points-txt d'une taille de six mille kilo-octets. 

+ datapath est la variable la plus importante 
  - elle vous indique l'emplacement du fichier téléchargé. 
  
+ Ce n'est pas l'emplacement du fichier sur l'ordinateur de l'utilisateur, c'est l'emplacement du fichier sur l'ordinateur qui exécute shiny, ce qui signifie que shiny peut accéder au fichier à partir de cet emplacement. 

+ Si vous voulez lire le fichier téléchargé, alors input-dollar-sign-input-id-dollar-sign-datapath est ce que vous utilisez. 

+ Par exemple, si shiny attend un fichier texte, vous pouvez utiliser la fonction readLines avec ce chemin de données comme premier argument.

+ Le premier argument de la fonction conditionalPanel est `condition`. 

+ La condition implique généralement la vérification ou la comparaison de la valeur d'une entrée. 

+ Lorsque vous écrivez la condition, si vous voulez faire référence à une valeur d'entrée, au lieu de input-dollar-sign-id, vous utilisez input-dot-id. 

+ Après le premier argument du panneau conditionnel, vous pouvez placer autant d'éléments d'interface utilisateur que vous le souhaitez. 

+ Il peut s'agir de texte, d'entrées, de sorties - tout ce qui se trouve dans fluidPage peut être placé ici. 

+ Ces éléments UI seront affichés lorsque la condition est remplie, et lorsque la condition n'est pas vraie, ces UI seront cachés.


####**6. Panneaux conditionnels**

###**Ajuster la réactivité**

#####**1. Ajuster la réactivité**
+ Passons en revue certaines des bases de la réactivité brillante.

####**2. Revue de la réactivité**
L+ es variables réactives, telles que les entrées ou les variables créées avec la fonction **reactive()**, sont différentes des variables ordinaires car elles peuvent automatiquement déclencher la réévaluation de tout code qui en dépend lorsqu'elles sont modifiées. 

+ Qu'est-ce que cela signifie que le code dépend d'une variable réactive ? 

+ En accédant simplement à une variable réactive, celle-ci devient une dépendance pour le code qui l'appelle. 

+ Par exemple, dans ce code qui définit une variable $x$, trois valeurs sont utilisées : 
        input$num1, input$num2 
et une variable réactive $y$. 

+ Le code pour évaluer $x$ sera ré-exécuté chaque fois que $y$ ou l'une des entrées est modifiée.

####**3. Isoler**
+ Ce comportement peut être contrôlé avec la fonction isolate. 

+ Isoler un morceau de code signifie que tout ce qui se trouve dans ce code ne créera pas de dépendance réactive. 

+ En d'autres termes, normalement, lorsqu'une valeur réactive change, le bloc de code entier est réexécuté, mais si une valeur réactive située à l'intérieur d'un isolat est modifiée, le code n'est pas réexécuté. 

+ Vous pouvez considérer l'isolate comme un moyen d'échapper aux déclenchements automatiques de la réactivité. Dans ce code, 
        input$num1 
      
a maintenant été isolé, donc maintenant si 

        input$num1 
change, rien ne se passera. 

+ Ce n'est que lorsque 
      $y$ ou input$num2 
      
change, que 
        $x$ 
sera réévalué. 

+ Vous pouvez également isoler plus d'une seule variable à la fois, vous pouvez mettre n'importe quel code à l'intérieur d'une fonction d'isolation. 

+ Ici, la multiplication de 

        input$num1 
avec 
        input$num2 
a été isolée. 

+ Le résultat est que maintenant $x$ ne sera réévalué que lorsque $y$ sera mis à jour, mais pas lorsque l'une des entrées le sera.

####**4. Isolez tout**
+ Et si, au lieu de réévaluer $x$ lorsque l'une de ses dépendances réactives est mise à jour, vous vouliez que l'utilisateur décide de la mise à jour de $x$ ? 

+ Dans ce cas, vous voudriez isoler toutes les variables réactives, de sorte que rien ne se passe lorsqu'une variable réactive est mise à jour. 

+ Mais il resterait le problème de l'exécution du code réactif $x$ à la demande.

####**5. Boutons d'action**
+ C'est ici que les boutons d'action peuvent être utiles. 

+ Les boutons d'action (ou, plus simplement, les boutons) sont considérés comme des entrées 

    - ils ont un inputId et un label comme toutes les autres entrées. 
    
+ Mais ils se comportent un peu différemment des autres entrées. La plupart des entrées fournissent à l'utilisateur un widget dont il peut modifier la valeur, et cette valeur devient accessible dans le serveur. 

+ Les boutons n'ont qu'une seule interaction possible : 

  + l'utilisateur peut cliquer dessus. 
  
+ Comme un bouton n'a pas de valeur intrinsèque, la valeur de l'entrée du bouton est un nombre entier. 

+ Chaque fois qu'un bouton est cliqué, sa valeur est incrémentée de un.


####**6. Les boutons d'action comme déclencheurs de réactivité**

+ Cela signifie que les boutons peuvent être utiles pour déclencher la réactivité. 

+ Comment ? Eh bien, en accédant simplement à la valeur d'un bouton, avec input$button-id, le bouton devient une dépendance réactive.

+ Pour continuer notre exemple de calcul de $x$, afin d'utiliser un bouton pour contrôler la réactivité, la première étape serait d'ajouter un bouton à l'interface utilisateur. 

+ Donnons-lui l'identifiant calculate-underscore-$x$. 

+ Ensuite, dans le code qui calcule $x$, 

        input-dollar-sign-calculate-$x$ 
        
doit être utilisé en dehors d'un **isolat**. 
      
+ Cette valeur d'entrée ne doit pas faire partie d'une expression, elle peut simplement se trouver sur sa propre ligne. 

+ C'est suffisant pour en faire une dépendance. 

+ Maintenant, chaque fois que le bouton est cliqué, sa valeur est incrémentée, ce qui entraîne le recalcul de $x$.

````r
---
title: "DASHBOARD"
output: 
  flexdashboard::flex_dashboard:
    orientation: rows
    source_code: embed
    vertical_layout: scroll
    html_document:
      df_print: paged
---


```{r, echo = TRUE, include = FALSE, message = FALSE}
library(flexdashboard)
library(plyr)
library(quantmod)
library(ggplot2)
library(dplyr)
library(DT)
library(pdfetch)
library(lubridate)
library(dygraphs)
library(tidyr)
library(PerformanceAnalytics)
library(stocks)
library(xts)
library(knitr)
library(symbols)
```


# Conclusion {.sidebar}

 

* This dashboard shows the analysis done for Tesla, Amazon, Netflix and General Electric.

* From the dashboard we can see that general electric had the lowest closing while amazon the highest.

* For long term investment amazon looks a good bet.

* Monthly return of tesla is volatile while amazon is mostly stable.

* For short-term investment tesla looks the best option as it has the highest stock price and its moving average is lower than the closing price.



# Instructions


Row {.tabset}
-----------------------------------------------------------------------

Overview

Due to the inherent growth in the electronic production and storage of information, there is often a feeling of “information overload” or inundation when facing the process of quantitative decision making. As an analyst your job will often be to conduct analyses or create tools to support quantitative decision making.

A principle tool used in industry, goverment, non-profits, and academic fields to compensate for the information overload is the information dashboard. Functionally, a dashboard is meant to provide a user with a central resource to present in a clear and concise manner all the information neccessary to support day-to-day decision making and support operations.

Objective

The objective of this laboratory is to plan, design, and create an information dashboard to support quantitative decision making. To accomplish this task you will have to complete a number of steps:

* Delineate the necessary decision (I will do that below).

* Identify what information will be relevant to decision making.

* Find and collect the data necessary to create your visualization plan.

* Organize and summarize the collected data.

* Design and create the best visualizations to present that information.

* Finally organize the layout of those visualizations in a way that conforms to the theory of dashboarding.

* Write a summary about what decisions you made based on the visualizations that you developed.



# Key Indicator Analysis


Row {.tabset}
-----------------------------------------------------------------------


* Here we will look after some key ratios extremely important in understanding the stock movement.

* Price/Earnings ratio, Price/EPS estimate for next year, dividend yield, market capitalization, moving average for 50 days, moving average for 200 days and average daily trade volume are included to reach a decision. 

* Some indicators helped in understanding the overall position of the company while the others showed the probablity of profit in short term investment. 

* In order to gain profit from the short term investment, we will invest in tesla.


```{r}
#Ratios needed for analysis
ratios <- yahooQF(c("P/E Ratio","Price/EPS Estimate Next Year","Dividend Yield","Market Capitalization","50-day Moving Average","200-day Moving Average","Average Daily Volume"))

tick <- c("TSLA","AMZN","NFLX","GE")

#Get the data
metrics <- getQuote(paste(tick, sep="", collapse=";"), what = ratios)

#Replacing date stamps to tick as first column
metrics <- data.frame(Symbol=tick, metrics[,2:length(metrics)])

DT::datatable(metrics)
```



Comparison between stock prices from June 17, 2010 to June 17, 2020.

------------------------------------

Column {.tabset}
------------------------------------
  
  
```{r}
getSymbols("TSLA",from="2010-06-17",to="2020-06-17")
TSLA_log_returns<-TSLA%>%Ad()%>%dailyReturn(type='log')

getSymbols("AMZN",from="2010-06-17",to="2020-06-17")
AMZN_log_returns<-AMZN%>%Ad()%>%dailyReturn(type='log')

getSymbols("NFLX",from="2010-06-17",to="2020-06-17")
NFLX_log_returns<-NFLX%>%Ad()%>%dailyReturn(type='log')

getSymbols("GE",from="2010-06-17",to="2020-06-17")
GE_log_returns<-GE%>%Ad()%>%dailyReturn(type='log')
```

# Visual Representation 1


Row {.tabset}
-----------------------------------------------------------------------


TESLA



```{r}
TSLA%>%Ad()%>%chartSeries(theme = chartTheme("white"),color.vol = TRUE, multi.col = FALSE)
TSLA%>%chartSeries(TA='addBBands();addVo();addMACD()',subset='2020',up.col = "Blue", dn.col = "Orange",theme = chartTheme("white"),color.vol = TRUE, multi.col = FALSE)
```

---------------------------------------

AMAZON



```{r}
AMZN%>%Ad()%>%chartSeries(theme = chartTheme("white"),color.vol = TRUE, multi.col = FALSE)
AMZN%>%chartSeries(TA='addBBands();addVo();addMACD()',subset='2020',up.col = "Blue", dn.col = "Orange",theme = chartTheme("white"),color.vol = TRUE, multi.col = FALSE)
```


---------------------------------------

NETFLIX


```{r}
NFLX%>%Ad()%>%chartSeries(theme = chartTheme("white"),color.vol = TRUE, multi.col = FALSE)
NFLX%>%chartSeries(TA='addBBands();addVo();addMACD()',subset='2020',up.col = "Blue", dn.col = "Orange",theme = chartTheme("white"),color.vol = TRUE, multi.col = FALSE)
```

---------------------------------------

GENERAL ELECTRIC


```{r}
GE%>%Ad()%>%chartSeries(theme = chartTheme("white"),color.vol = TRUE, multi.col = FALSE)
GE%>%chartSeries(TA='addBBands();addVo();addMACD()',subset='2020',up.col = "Blue", dn.col = "Orange",theme = chartTheme("white"),color.vol = TRUE, multi.col = FALSE)
```


# Visual Representation 2


Row {.tabset}
-----------------------------------------------------------------------





```{r}
tick1 <- c("TSLA", "AMZN", "NFLX", "GE")
ClosingPrice <- do.call(merge, lapply(tick1, function(x) Cl(get(x))))

dateperiod<-c("2010-06-17", "2020-06-17")


dygraph(ClosingPrice, main="Closing Price in USD", group="Stock") %>%
  dyAxis("y", label="Closing Price(USD)") %>%
  dyOptions( colors = RColorBrewer::brewer.pal(5, "Set2")) %>%
  dyHighlight(highlightSeriesBackgroundAlpha = 0.5,
              highlightSeriesOpts = list(strokeWidth = 4)) %>%
  dyRangeSelector(height = 30)

```





---------------------------------------





```{r}
m.rt.TSLA <- monthlyReturn(TSLA)
m.rt.AMZN <- monthlyReturn(AMZN)
m.rt.NFLX <- monthlyReturn(NFLX)
m.rt.GE <- monthlyReturn(GE)

mg.return <- merge.xts(m.rt.TSLA,m.rt.AMZN, m.rt.NFLX, m.rt.GE)
colnames(mg.return) <- c('TSLA','AMZN','NFLX','GE')


dygraph(mg.return, main = "Monthly Return") %>%
  dyAxis("y", label = "Return") %>%
  dyOptions(colors = RColorBrewer::brewer.pal(5, "Set2")) %>%
  dyHighlight(highlightSeriesBackgroundAlpha = 0.5,
              highlightSeriesOpts = list(strokeWidth = 4)) %>%
  dyRangeSelector(height = 30)
```
````
