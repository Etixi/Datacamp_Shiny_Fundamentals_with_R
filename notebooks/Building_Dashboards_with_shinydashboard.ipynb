####**2. Intrants**
+ Rappelez-vous de votre introduction à shiny que vous pouvez créer des widgets pour prendre des entrées dans vos applications. 

+ Quelques exemples de ceux-ci sont les boutons d’action, un bouton dont la valeur initiale est 0 et incrémente chaque fois qu’il est cliqué, une case à cocher pour spécifier des valeurs logiques, une entrée de date, une entrée numérique, des boutons radio, sélectionner une entrée à sélectionner dans un menu déroulant, une entrée de curseur et une entrée de texte. 

+ Remarquez le modèle ici 
  - comme avec la plupart des fonctions brillantes, ce sont camelCase, avec le premier mot décrivant le type d’entrée, et le deuxième mot le plus souvent « Input » (bien que cela puisse varier, par exemple il peut être « Button » ou « Buttons » comme on le voit ici).

####**3. Intrants**
+ Chacune de ces entrées suit une forme similaire. 

+ Par exemple, la fonction selectInput() prend trois paramètres obligatoires : 
  + inputId -- c’est l’entrée que vous utiliserez pour accéder à la valeur qui a été sélectionnée ailleurs dans l’application, par exemple ici mon inputId est « caractère de soulignement favori » -- ce paramètre est commun à toutes les fonctions d’entrée, label -- l’étiquette d’affichage vue dans l’interface utilisateur de l’application, ici « Caractère favori », encore une fois, ce paramètre est également commun à toutes les fonctions d’entrée, et dans le cas de la fonction **selectInput()**, les choix - ici vous entrez les choix que vous souhaitez afficher, par exemple ici j’ai défini les choix sur « Leia » « Luke » et « Han ».

####**4. Fonctions de rendu**
+ Les fonctions de rendu résident généralement sur le serveur. Ils prennent un argument, une expression R. 

+ Voici quelques exemples de fonctions de rendu -- **renderPrint()** restitue n’importe quelle sortie imprimée, **renderText()** restitue les chaînes de caractères, **renderTable()** une trame de données, une matrice ou une table, **renderDataTable()** rend une table de données, **renderPlot()** un plot, **renderImage()** une image et **renderUI()** rend HTML ou un objet de balise brillant. 

+ Remarquez à nouveau le modèle camelCase, avec « render » toujours comme premier mot et la chose qui est rendue comme le second.

####**6. Fonctions de sortie**

###**Optimisation des performances**

####**1. Optimisation des performances**

+ Très bien, parlons brièvement de l’optimisation des performances de votre tableau de bord brillant. 

+ Il existe de nombreuses façons d’optimiser les performances, ici nous allons nous concentrer spécifiquement sur l’endroit où vous placez un certain code.

####**2. En dehors de la fonction serveur**
+ Le code situé en dehors de la fonction serveur ne sera exécuté qu’une seule fois. 

+ Par exemple, le code ici, l’ensemble de données Star Wars ne sera chargé qu’une seule fois lorsque l’application est lancée. 

+ Il s’agit de l’endroit approprié pour charger des ensembles de données statiques tels que celui-ci.

####**3. À l’intérieur de la fonction serveur**
+ Le code situé à l’intérieur de la fonction serveur, mais à l’extérieur d’un objet réactif s’exécutera chaque fois qu’un nouvel utilisateur visitera l’application. 

+ Cela peut être inefficace, car il vous suffit de charger ces données une seule fois.

####**4. À l’intérieur de la fonction serveur, à l’intérieur d’un objet réactif**
+ Le code situé à l’intérieur d’une fonction serveur ET à l’intérieur d’un objet réactif s’exécutera chaque fois que l’utilisateur modifiera l’entrée. 

+ Par exemple, nous créons ici un sous-ensemble de données Star Wars basé sur les entrées de l’utilisateur, indiquant la hauteur maximale des personnages dans notre nouvel ensemble de données. 

+ Chaque fois que l’utilisateur modifie l’entrée, les données sont rechargées. 

+ Dans ce cas particulier, puisque les mêmes données sont rechargées à chaque fois, c’est une inefficacité évidente, et une excellente occasion d’accélérer les performances d’une application !

####**1. Explication de Bootstrap**

+ Vous n’avez certainement pas besoin d’être un expert en la matière, mais je vais passer en revue quelques conseils qui rendront le processus de mise en page un peu plus intuitif.

####**2. Mise en page Bootstrap**

+ La disposition bootstrap du corps brillant du tableau de bord se compose de 12 colonnes de largeur égale et d’un nombre quelconque de lignes. 

+ Lorsque vous placez une boîte, vous pouvez définir le nombre de colonnes que vous souhaitez qu’elle prenne.

####**4. Types de mise en page**
+ Il existe trois types de dispositions de base : 
  + disposition basée sur les lignes, 
  + disposition basée sur les colonnes 
  +  mise en page mixte.

####**5. Mise en page basée sur les lignes**


+ Avec la disposition basée sur les lignes, vous enveloppez vos boîtes dans une fonction **fluidRow()** pour former chaque ligne. 

+ Le haut des cases de chaque rangée sera aligné. 

+ Vous pouvez aligner le fond des boîtes en définissant la hauteur dans chaque boîte pour qu’elle soit la même.

####**6. Mise en page basée sur les colonnes**

+ Avec la disposition basée sur les colonnes, vous utilisez la fonction colonne dans une ligne fluide pour définir les colonnes. 

+ Notez ici que nous définissons maintenant la largeur dans la fonction colonne (dans l’exemple que vous voyez, c’est toujours 6), mais nous définissons la largeur de la boîte sur « NULL » 

  - il est important de se rappeler lorsque vous utilisez une disposition basée sur des colonnes 
  - les largeurs sont toujours définies dans les fonctions de colonne.

####**7. Mise en page mixte**

+ Une disposition mixte est simplement une combinaison de la disposition des lignes et des colonnes. 

+ Par exemple, ici, j’ai une ligne avec une boîte qui occupe 12 colonnes, puis en dessous, j’ai une autre ligne avec des dispositions de colonnes.

####**Instructions**

+ Créez l'objet n_us indiquant combien de boules de feu ont atterri aux États-Unis. Vous pouvez utiliser le code fourni dans le premier chunk ci-dessus pour le faire.

+ Créez une boîte de valeur dynamique en utilisant renderValueBox() et valueBox() avec n_us comme valeur, "Number of Fireballs in the US" comme sous-titre, et l'icône "globe".

+ Ajoutez la couleur du paramètre à la boîte de valeur, si le nombre de boules de feu aux États-Unis est inférieur à 10, rendez-la "bleue", sinon rendez-la "fuchsia".

+ Relancez le tableau de bord brillant pour voir votre travail.

```r
n_us <- sum(
  ifelse(
    nasa_fireball$lat < 64.9 & nasa_fireball$lat > 19.5
      & nasa_fireball$lon < -68.0 & nasa_fireball$lon > -161.8,
    1, 0),
  na.rm = TRUE)
  
server <- function(input, output) {
  output$us_box <- renderValueBox({
    valueBox(
      value = n_us,
      subtitle = "Number of Fireballs in the US",
      icon = icon("globe"),
      color = if (n_us < 10) {
                "blue"
              } else {
              "fuchsia"
              }
    )
  })
}

body <- dashboardBody(
  fluidRow(
    valueBoxOutput("us_box")
  )
)
ui <- dashboardPage(header = dashboardHeader(),
                    sidebar = dashboardSidebar(),
                    body = body
                    )
shinyApp(ui, server)
```
